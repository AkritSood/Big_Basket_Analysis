```python
product_count = df['product'].value_counts()
top_selling=product_count.nlargest(10)
print(top_selling)
```

```python
least_selling=product_count.nsmallest(10)
print(least_selling)
```

```python
combined_sales = pd.concat([top_selling, least_selling])

plt.figure(figsize=(10, 8))
bars = plt.bar(range(len(combined_sales)), combined_sales.values)
plt.xlabel("Products")
plt.ylabel("Number of Items Sold")
plt.title("Top and Least Selling Products")
plt.xticks(range(len(combined_sales)), combined_sales.index, rotation=45, ha='right')

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(i, height, str(int(height)), ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
df['discount_amount'] = df['market_price'] - df['sale_price']
df['discount_percent'] = ((df['discount_amount']) / df['market_price']) * 100

df_sort=df.sort_values(by='discount_percent', ascending=False)
highest_discounts = df_sort.head(10)
lowest_discounts = df_sort.tail(10)
```

```python
print("Top 10 Discounts:")
print(highest_discounts[['product','market_price', 'sale_price', 'discount_percent']])

print("\Least 10 Discounts:")
print(lowest_discounts[['product','market_price', 'sale_price', 'discount_percent']])
```

```python
df = df.dropna(subset=['product'])
df = df.dropna(subset=['brand'])
```

```python
df['sale_price'] = df['sale_price'].fillna(df['sale_price'].mean())
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
df['description'] = df['description'].fillna('Unknown')
```

```python
df['discount_amount'] = df['discount_amount'].fillna(df['discount_amount'].mean())
df['discount_percent'] = df['discount_percent'].fillna(df['discount_percent'].mean())
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(2, 2, 1)
sns.boxplot(y=df['sale_price'])
plt.title('Sale Price Box Plot')

plt.subplot(2, 2, 2)
sns.boxplot(y=df['market_price'])
plt.title('Market Price Box Plot')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['discount_amount'])
plt.title('Discount Amount Box Plot')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['rating'])
plt.title('Rating Box Plot')

plt.tight_layout()
plt.show()
```

```python
Q1 = df['sale_price'].quantile(0.25)
Q3 = df['sale_price'].quantile(0.75)
IQR = Q3 - Q1
print(f'''Q1 is : {Q1}, Q3 is {Q3}, IQR is {IQR}.''')

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'''Lower Bound is : {lower_bound}, Upper Bound is {upper_bound}.''')

outliers = df[(df['sale_price'] < lower_bound) | (df['sale_price'] > upper_bound)]

mean_sale_price = df['sale_price'].mean()
df.loc[outliers.index, 'sale_price'] = mean_sale_price
```
